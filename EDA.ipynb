```python
import os 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import cv2
from Utils.DataExtractor import DataExtractor
```

```python
DATA_DIR = os.path.join(os.getcwd(), "data")
print(os.listdir(DATA_DIR))
```

```python
data_extractor = DataExtractor

train_path = os.path.join(DATA_DIR, "train.csv")

train_metadata = data_extractor.read_csv(fpath=train_path)
if train_metadata.empty:
    print("EMPTY DATA SET")
print(train_metadata.head())
```

```python
print(" Number of entries : ", train_metadata.shape)
```

```python
print("Number of unique IDs provided :", train_metadata["Id"].nunique())
```

```python
idgroup = train_metadata.groupby(by="Id", as_index=False).count()
idgroup.columns = ["Id", "count"]
idgroup.sort_values(by="count", inplace=True, ascending=False)
idgroup.reset_index(drop=True)
```

```python
# Ignore new_whale and plot each class
image_dir = os.path.join(DATA_DIR, "train")
```

```python
counter = 0
idlist = idgroup["Id"].values
for id in idlist:
    if id == "new_whale":
        continue
    # Get a list of all the images for this class
    image_list = train_metadata[train_metadata["Id"] == id]["Image"].values
    print("------------------------------------------------------------------------------------------------------------ ")
    fig = plt.figure(figsize=(20,8))
    nrows = 2
    ncols = 4
    for i in range(min(nrows*ncols, len(image_list))):
        image = image_list[i]
        img_path = os.path.join(image_dir, image)
        ax1 = fig.add_subplot(nrows,ncols,i+1)
        img = cv2.imread(img_path)
        ax1.imshow(img)
#         break
    fig.suptitle(id, size=25)
    plt.show()
    
    counter += 1
    if counter > 5:
        break
```
